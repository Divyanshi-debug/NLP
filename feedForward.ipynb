```python
import numpy as np
```

```python
x = np.array([
    [1,0,1,0],
    [1,0,1,1],
    [0,1,0,1]])

y = np.array([[1], [1], [0]])
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
inputNeurons = x.shape[1]
hiddenNeuron = 3
outputNeurons = 1
```

```python
weightsHidden = np.random.uniform(size=(inputNeurons, hiddenNeuron))
biasHidden  = np.random.uniform(size=(1, hiddenNeuron))
weightsOutput = np.random.uniform(size=(hiddenNeuron, outputNeurons))
biasOutput = np.random.uniform(size=(1, outputNeurons))
```

```python
fx = np.dot(x,weightsHidden) + biasHidden
```

```python
hiddenLayer = sigmoid(fx)
hiddenLayer
```

```python
fx_ = np.dot(hiddenLayer, weightsOutput) + biasOutput
```

```python
outputLayer = sigmoid(fx_)
outputLayer
```
