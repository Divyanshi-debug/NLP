```python
import nltk
import string
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize #tokenization
from nltk.corpus import stopwords #
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
from typing_extensions import final
sentences=[
    "i am learning python and python",
    "we are happy today",
    "i am sad today"
]

def preprocess_text(documents):
  cleaned=[]
  for doc in documents:
    raw_text=doc.lower()
    token=word_tokenize(raw_text)
    english_stopwords=stopwords.words("english")
    filtered=[]
    for word in token:
      if word not in english_stopwords:
        filtered.append(word)
    clear_tokens=[word for word in filtered if word not in string.punctuation]
    lemmatize_words=[]
    for word in clear_tokens:
      lemmatize_words.append(WordNetLemmatizer().lemmatize(word,"v"))
    final_words=[]
    for word in lemmatize_words:
      if word.isalpha():
        final_words.append(word)
    cleaned_text=" ".join(final_words)
    cleaned.append(cleaned_text)
  return cleaned




cleaned_sentences=preprocess_text(sentences)

print(cleaned_sentences)

```
