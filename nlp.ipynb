**NLP PHASE 1**

```python
#TEXT PROCESSING PIPELINE
#text->cleaning->tokenization->stopwords removal->stemming/
```

```python
text="i am learning Python programming and Python is fast......"
print("Original text: ",text)
```

```python
#Step-1: convert to lowercase
text=text.lower()
#Python!=python
print("Lowercase text: ",text)
```

```python
#Step-2: Remove punctuation- !@#$%^&*()_;'.,
#re-regex-regular expression
import re
text=re.sub(r'[^\w\s]','', text)
#remove everything except letters and spaces
#path="C:\Gen-AI\chat_app": will give error as it treats \ with words as.........
#path="C:\\Gen-AI\\chat_app": will work
#python soln: path=r"C:\Gen-AI\chat_app" will work
print("Without punctuation:", text)
```

```python
#Step-3: Tokenization
tokens=text.split()
print("Tokens: ",tokens)
```

```python
#Step-4: Remove Stopwords- is, am, are, the, a, an, i
stopwords=["is","am","are","the","a","an","i","and"]
filtered_tokens=[]
for word in tokens:
    if word not in stopwords:
        filtered_tokens.append(word)
print("After removing stopwords: ",filtered_tokens)
```

```python
def stem(word):
  if word.endswith("ing"):
    return word[:-3]
  return word

stemmed_words=[]
for word in filtered_tokens:
  stemmed_words.append(stem(word))
print("After stemming :",stemmed_words)
```

**NLP PHASE 2**

```python
# Why vectors?
# ML models only understand numbers

#Bag of Words (BoW)
```

```python
#Will convert text->count
from sklearn.feature_extraction.text import CountVectorizer
```

```python
sentences=[
    "i am learning python and python",
    "we are happy today",
    "i am sad today"
]
```

```python
vectorizer=CountVectorizer() #create object
X=vectorizer.fit_transform(sentences)
#learns vocabulary
#convert sentences to numbers
```

```python
print("Vocabulary: ",vectorizer.get_feature_names_out())
#1. all unique words
#2. sorted in alphabetical order
#3. remove single letters like i,a,
```

```python
print("Vectors:")
print(X.toarray())
#Each row is a sentence
#Each column is a word
#Each value is a count
```

**TF-IDF=Term Frequency Inverse Document Frequency**

```python
#Why TF-IDF?
#Because count is not enough
#Bag of words treats all words equally
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences=[
    "i am learning python and python",
    "we are happy today",
    "i am sad today"
]
```

```python
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(sentences)
#build vocabulary
#calculate TF
#calculate idf-impact of each word
#multiplies-tf x idf
#produces matrix
```

```python
print("Vocabulary :",vectorizer.get_feature_names_out())
```

```python
print("TF-IDF Matrix:")
print(X.toarray())
```

```python
#count+impact of each word
#TF-IDF rewares rare important words and penalizes common words
#TF=count(words)/total_words
#IDF=log(total_docs/documents_containing word)
```
