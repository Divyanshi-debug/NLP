```python
from nltk.tokenize import word_tokenize #tokenization
from nltk.corpus import stopwords #
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
raw_text="I am learning Python programming, Python is the greatest language.........!!"
```

```python
#step-1: lowercase
raw_text=raw_text.lower()
print("After lowercase:",raw_text)
```

```python
#step-2:tokenization
print(raw_text.split())
```

```python
tokens=word_tokenize(raw_text)
print("Tokens:",tokens)
```

```python
#step-3:remove stopwords
english_stopwords=stopwords.words("english")
print("Stopwords:",english_stopwords)
```

```python
filtered_tokens=[]
for word in tokens:
    if word not in english_stopwords:
        filtered_tokens.append(word)
print("Filtered tokens: ",filtered_tokens)
```

```python
#step-4: remove punctuations
import string
punctuations=string.punctuation
print("Punctuations available:",punctuations)
```

```python
clear_tokens=[word for word in filtered_tokens if word not in punctuations]
print("After removing punctuations: ",clear_tokens)
```

```python
stem=PorterStemmer()
#stem.stem("playing")
#stem.stem("watching")
#stem.stem("played")
#stem.stem("flying")
stem.stem("bought")
```

```python
wnet=WordNetLemmatizer()
wnet.lemmatize("playing","v")
#wnet.lemmatize("watching","v")
#wnet.lemmatize("played","v")
#wnet.lemmatize("bought","v")
wnet.lemmatize("went","v") #v:verb
wnet.lemmatize("greatest","a")
```

```python
lemmatized_words=[]
for word in clear_tokens:
    lemmatized_words.append(wnet.lemmatize(word,"v"))

print("After Lemmatization :",lemmatized_words)
```

```python
final_tokens=[]
for word in lemmatized_words:
    if word.isalpha():
      final_tokens.append(word)

print("Final Tokens:", final_tokens)
```

```python
cleaned_text=" ".join(final_tokens)
print("Cleaned text:",cleaned_text)
```
