**NLP BASED CHATBOT**

```python
#tokenization
from nltk.tokenize import word_tokenize
#stopwords list
from nltk.corpus import stopwords
#stemming
from nltk.stem import PorterStemmer
#lemmatization
from nltk.stem import WordNetLemmatizer
import string
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
```

```python
nltk.download("punkt")
nltk.download("stopwords")
nltk.download("wordnet")
nltk.download('punkt_tab')
```

```python
training_data=[
    ("hello","greet"),
    ("hi","greet"),
    ("hello there","greet"),
    ("hi there","greet"),
    ("hey","greet"),
    ("good morning","greet"),
    ("what's the weather today","weather"),
    ("how's the weather today","weather"),
    ("what's the temperature","weather"),
    ("is it raining","weather"),
    ("tell me temperature","weather"),
    ("open google","open_web"),
    ("open facebook","open_web"),
    ("open youtube","open_web"),
    ("go to google","open_web"),
    ("bye","exit"),
    ("goodbye","exit"),
    ("see you later","exit"),
    ("exit","exit")
    ]
```

```python
sentences=[] #features or x
labels=[] #target or y

for text,intent in training_data:
    sentences.append(text)
    labels.append(intent)
```

```python
def preprocess_text(documents):
  cleaned=[]
  for doc in documents:
    raw_text=doc.lower()
    token=word_tokenize(raw_text)
    english_stopwords=stopwords.words("english")
    filtered=[]
    for word in token:
      if word not in english_stopwords:
        filtered.append(word)
    clear_tokens=[word for word in filtered if word not in string.punctuation]
    lemmatize_words=[]
    for word in clear_tokens:
      lemmatize_words.append(WordNetLemmatizer().lemmatize(word,"v"))
    final_words=[]
    for word in lemmatize_words:
      if word.isalpha():
        final_words.append(word)
    cleaned_text=" ".join(final_words)
    cleaned.append(cleaned_text)
  return cleaned




#cleaned_sentences=preprocess_text(sentences)
```

```python
sentences[:5]
```

```python
cleaned_data=preprocess_text(sentences)
```

```python
cleaned_data
```

```python
from numpy import vectorize
vectorizer=TfidfVectorizer()
X=vectorizer.fit_transform(cleaned_data)
```

```python
model=LogisticRegression()
model.fit(X,labels)
```

```python
#fit: make vocabulary, needs 2 arguments, x and y, model trains
#transform: fit testing data into vocabulary, applies maths, use transform to avoid re-vocabulary
#fit transform: used for training data, perform 2 steps: fit and transform, i.e, train and apply maths
#predict: pass testing input, it will predict
```

```python
user_msg="open LinkedIn"
processed=preprocess_text([user_msg])
user_vector=vectorizer.transform(processed) #use transform, not fit transform
prediction=model.predict(user_vector)
print("prediction is: ",prediction)
```
