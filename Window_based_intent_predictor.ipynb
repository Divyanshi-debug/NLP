```python
import gensim.downloader as api
import numpy as np
from nltk.tokenize import word_tokenize
import nltk
nltk.download("punkt_tab")
```

```python
print("Loading model......")
model=api.load("glove-wiki-gigaword-100")
print("Loaded")
```

```python
sentence="Please open google and check weather"
tokens=word_tokenize(sentence.lower())
```

```python
#Sliding window function
def get_windows(tokens_list,window_size=2):
  windows=[]
  for i in range(len(tokens)):
    left=max(0,i-window_size)
    right=min(len(tokens),i+window_size+1)
    windows_words=tokens[left:right]
    windows.append(windows_words)
  return windows
```

```python
windows=get_windows(tokens,2)
print("Windows:",windows)
```

```python
# 1st word: please: 2 words from left and 2 words from right
# next word: open: 2 words from left and 2 words from right
# and it moves like this till last word: weather: 2 words from left and 2 words from right
```

```python
#Next Step: converting window to embedding
def window_vector(window):
  vectors=[]
  for word in window:
    if word in model:
      #get embedding of a word from the glove model
      vectors.append(model[word])

  if len(vectors)==0:
    return np.zeros(100)

  return np.mean(vectors, axis=0)
```

```python
windows_vector=[window_vector(w) for w in windows]
```
